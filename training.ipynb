```python
label_encoder = preprocessing.LabelEncoder()
df['label'] = label_encoder.fit_transform(df['class'])
```

```python
dataset = df[['text', 'label']]
```

```python
dataset.info()
```

```python
import re

def preprocess_text(text):

  if not isinstance(text, str):
    return 'None'


  text = text.lower()
  text = re.sub(r'http\S+', '', text)

  space_pattern = r'\s+'
  text = re.sub(pattern=space_pattern, repl=" ", string=text)

  return text

# Apply the preprocessing function to each text in the list
dataset.loc[:, 'text'] = dataset['text'].apply(preprocess_text)
```

```python
dataset['text'] = dataset['text'].apply(lambda x: preprocess_text(x))
```

```python
dataset.dropna(inplace=True)
```

```python
text = dataset['text'].to_list()
label = dataset['label'].to_list()
```

```python
train_text, test_text, train_label, test_label = train_test_split(text, label, test_size=0.2, random_state=42)
```

```python
test_label
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
model_name = "distilbert/distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
class DepressionDataset(torch.utils.data.Dataset):
    def __init__(self, text, labels, tokenizer, max_len = 512):
        self.text = text
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __getitem__(self, idx):
        text = str(self.text[idx])
        label = torch.tensor(self.labels[idx])
        encoding = self.tokenizer(text, truncation=True, padding='max_length', max_length=self.max_len, return_tensors='pt')

        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': label
        }
    def __len__(self):
        return len(self.labels)
```

```python
train_dataset = DepressionDataset(train_text, train_label, tokenizer)
test_dataset = DepressionDataset(test_text, test_label, tokenizer)
```

```python
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
from sklearn.metrics import accuracy_score, f1_score

def compute_metrics(logits_and_labels):
    logits, labels = logits_and_labels
    preds = np.argmax(logits, axis=-1)
    accuracy = accuracy_score(labels, preds)
    f1 = f1_score(labels, preds, average='macro')

    return {
        'accuracy': accuracy,
        'f1': f1,

    }
```

```python
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2, ignore_mismatched_sizes=True)
```

```python
training_args = TrainingArguments(
    output_dir='./sucidal-text-classification-distillbert',
    eval_strategy="steps",
    num_train_epochs=1,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    learning_rate=2e-5,
    warmup_steps=500,
    weight_decay=0.01,
    eval_steps=5000,
    load_best_model_at_end=True,
    logging_strategy = "steps",
    logging_steps = 5000,
    save_strategy = "steps",
    save_steps = 5000,
    push_to_hub=True,
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)
```

```python
import torch
torch.cuda.empty_cache()
```

```python
trainer.train()
```

```python
eval_result = trainer.evaluate(eval_dataset=test_dataset)
```
